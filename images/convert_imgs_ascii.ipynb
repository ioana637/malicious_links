```python
import pandas as pd

def load_dataset_pakhare():
    file="../data/pakhare_urls.csv"
    df = pd.read_csv(file)
    # count
    print("Good samples:" + str(df.groupby('Class').size()['good']))
    print("Bad samples:" + str(df.groupby('Class').size()['bad']))
    print(df.info(verbose=False))
    # check for null values
    bool_series = pd.isnull(df["URLs"])
    print(df[bool_series])
    return df

df = load_dataset_pakhare()
```

```python
df_sample_10000_good = df[df['Class']=='good'].sample(n=10000)
df_sample_10000_bad = df[df['Class']=='bad'].sample(n=10000)
print(df_sample_10000_good)
print(df_sample_10000_bad)
```

```python
df = pd.concat([df_sample_10000_good, df_sample_10000_bad], axis=0)
print(df)
```

```python
from PIL import Image
import PIL
import os
import glob

folder_for_resized_images_benign = 'D:\IdeaProjects\malicious_links\data\images\pakhare_dataset\\resized_images\\benign'
folder_for_resized_images_malicious = 'D:\IdeaProjects\malicious_links\data\images\pakhare_dataset\\resized_images\\malicious'
folder_for_initial_images_benign = 'D:\IdeaProjects\malicious_links\data\images\pakhare_dataset\\benign'
folder_for_initial_images_malicious = 'D:\IdeaProjects\malicious_links\data\images\pakhare_dataset\\malicious'

import cv2
import os
def load_images_from_folder(folder):
    images = []
    for filename in os.listdir(folder):
        img = cv2.imread(os.path.join(folder,filename))
        if img is not None:
            images.append((img, filename))
    return images

images_benign = load_images_from_folder(folder_for_initial_images_benign)
images_malicious = load_images_from_folder(folder_for_initial_images_malicious)

max_dim_for_ascii_images = max([len(image[0]) for image in images_malicious + images_benign])
print(max_dim_for_ascii_images)
```

```python
def resize_images_for_folder_and_save(initial_folder, save_folder, resize_dimensions):
    directory_files = os.listdir(initial_folder)
    multiple_images = [file for file in directory_files if 'ascii_img' in file and file.endswith(('.png'))]
    print(multiple_images)
    for image in multiple_images:
        img = Image.open(initial_folder + '\\'+ image)
        imgResized = img.resize(size=resize_dimensions)
        print(imgResized)
        # We would run the command below to save the images:
        imgResized.save(save_folder + '\\'+ image, optimize=True)
        
```

```python
resize_dim = (max_dim_for_ascii_images, max_dim_for_ascii_images)
print(resize_dim)
```
