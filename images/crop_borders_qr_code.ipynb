```python
folder_with_qr_codes = 'D:\IdeaProjects\malicious_links\data\qr_images\pakhare_dataset\\resize_200_200'

import cv2
import os
def load_images_from_folder(folder):
    images = []
    for filename in os.listdir(folder):
        img = cv2.imread(os.path.join(folder,filename))
        if img is not None:
            images.append((img, filename))
    return images
```

```python
images = load_images_from_folder(folder_with_qr_codes)
```

```python
from PIL import Image, ImageChops

def trim(im):
    bg = Image.new(im.mode, im.size, im.getpixel((0,0)))
    diff = ImageChops.difference(im, bg)
    diff = ImageChops.add(diff, diff)
    bbox = diff.getbbox()
    if bbox:
        return im.crop(bbox)

def remove_border_qr_images_for_folder_and_save(initial_folder, save_folder =''):
    directory_files = os.listdir(initial_folder)
    multiple_images = [file for file in directory_files if 'qr_code' in file and file.endswith(('.png'))]
    print(multiple_images)
    
    for image in multiple_images[1:3]:
        img = Image.open(folder_with_qr_codes + '\\'+ image)
        print(img)
        img = trim(img)
        img.show()
        # We would run the command below to save the images:
        # img.save(save_folder + '\\'+ image, optimize=True)
```
